```python
import pandas as pd
data = pd.read_csv('./dataset.csv')
```

```python
data.head()
```

```python
x = data['celsius']
y = data['fahrenheit']
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), # celcius, one entry
    tf.keras.layers.Dense(units=1), #  la salida, los F
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss="mean_squared_error"
)
```

```python
training = model.fit(x, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoc")
plt.ylabel("Magnitud de perdida")

plt.plot(training.history['loss'])
```

```python
import numpy as np
# Reshape the input data to match the model's input shape
input_data = np.array([120])  # Convert to numpy array
prediction = model.predict(input_data.reshape(-1, 1)) # re sahe to the neuronal network input
prediction[0]
```

```python
print(f"Prediction: {prediction[0]}")
```
